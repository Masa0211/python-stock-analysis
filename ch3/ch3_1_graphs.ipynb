# Graphs

### Plot

```python
from matplotlib import pyplot as plt
import numpy as np

values = np.array([1, 2, 3, 4, 5])
height = np.array([100, 50, 300, 250, 500])
plt.plot(values, height)
plt.show()
```

### bar, pie, hist, scatter

```python
values = np.array([1, 2, 3, 4, 5])
height = np.array([100, 50, 300, 250, 500])
plt.bar(values, height)
plt.show()
```

```python
values = np.array([0.1, 0.05, 0.30, 0.25, 0.5])
names = ["A", "B", "C", "D", "E"]
plt.pie(values, autopct="%d%%", labels=names)
plt.axis("equal")
plt.show()
```

```python
values = np.random.normal(50, 10, 1000)
plt.hist(values)
plt.show()
```

```python
x = np.linspace(50,10,100)
y = np.random.randn(100)
plt.scatter(x,y)
plt.show()
```
